Installation Steps (for Ubuntu/Debian-based WSL)
Launch your WSL terminal (e.g., Ubuntu).

Update the package index to ensure you get the latest version information:

sudo apt update
You may be prompted for your user password.

Install pip for Python 3 (which is the standard version):

sudo apt install python3-pip
When prompted to continue, type Y and press Enter.

Note: If you are using an older version of Ubuntu or if you specifically need the Python 2 version (which is strongly discouraged), you would use python-pip, but for modern development, always use python3-pip.
Verify the installation: After the installation completes, check the installed version:

pip3 --version
This command should display the version of pip you just installed.

While you can now install packages globally using pip3 install <package_name>, it's highly recommended to use virtual environments. Virtual environments isolate your project dependencies, preventing conflicts between different projects and keeping your system-wide Python clean.

To create and use a virtual environment, follow these steps after installing python3-pip:

Install python3-venv (usually installed with python3-pip, but good to check):

sudo apt install python3-venv
Navigate to your project directory.

Create a virtual environment (named .venv in this example):

python3 -m venv .venv
Activate the virtual environment:

source .venv/bin/activate
Your terminal prompt should now show the name of your environment (e.g., (.venv)), indicating it's active.

Install packages using pip (not pip3 or sudo):

pip install numpy pandas
When the virtual environment is active, using the simple pip command will automatically use the pip associated with that environment and install packages locally within it.
